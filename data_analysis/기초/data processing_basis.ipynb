### 조건에 맞는 행 선택
---
- .ge = 이상, .gt = 초과, .le = 이하, .lt = 미만, .eq = 같다, .ne = 같지 않다<br>
- 출력값 -> true / false

### 1) 연속형 범주

```python
# 최소연봉 col 중 5000 이상인 값 반환
emp['최소연봉'].ge(5000)
```

```python
# 데이터프레임으로 반환
emp[emp['최소연봉'].ge(5000)]
```

```python
# 5000 <= 최소연봉 < 10000
emp[emp['최소연봉'].ge(5000) & emp['최소연봉'].lt(10000)]
```

```python
# 최소연봉 > 10000, 최대연봉 > 10000
emp[emp['최소연봉'].ge(10000) & emp['최대연봉'].ge(10000)]
```

### 2) 범주형 변수

```python
# (경력분야 = 신입) 데이터 추출
emp[emp['경력분야'].eq('신입')]
```

```python
# 경력구분 = 신입 or 인턴
emp[emp['경력분야'].eq('신입') | emp['경력분야'].eq('인턴')]
```

```python
# 원본 emp에 재할당
emp = emp[emp['경력분야'].isin(['경력', '신입', '인턴'])]
```

### 데이터프레임 병합
---

```python
ent.head()
```

```python
# 선택할 열
cols = ['기업일련번호', '업체국문명', '기업소개상세내용', '설립일자', 
        '구성원수', '누적투자유치금액', '기업추천수', '산업구분명']
```

### 외래키 확인

```python
ent.dtypes
```

```python
# ent = ent.astype({'기업일련번호': 'str'})
ent = ent.rename(columns = {'기업일련번호': '기업번호'})  # 열 이름 변경
```

### emp, ent 병합
#### 1) 외래키 값이 일치하는 갯수

```python
len(set(emp['기업번호']) & set(ent['기업번호']))
```

```python
ent['기업번호'].duplicated().sum()
```

```python
# how = 병합방법 ('inner', 'outer', 'left')
# on = 외래키 지정
mgd = pd.merge(left = emp, right = ent, how = 'inner', on = '기업번호')
```

```python
mgd = mgd.sort_values(by=['경력분야', '연봉구분'])
mgd.head()
```

```python
# csv 파일로 저장
mgd.to_csv(path_or_buf = 'Merged_Data.csv', index=None)
```
